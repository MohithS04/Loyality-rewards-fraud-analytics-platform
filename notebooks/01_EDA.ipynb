# Exploratory Data Analysis (EDA)

This notebook performs basic exploratory data analysis on the Loyalty Rewards Fraud dataset.

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')

# Load Data
members = pd.read_csv('../data/members.csv')
redemptions = pd.read_csv('../data/redemptions.csv')
```

## Data Overview

```
print(f"Members: {members.shape}")
print(f"Redemptions: {redemptions.shape}")

print("\nFraud Distribution:")
print(redemptions['is_fraud'].value_counts(normalize=True) * 100)

print("\nFraud Types:")
print(redemptions[redemptions['is_fraud']==1]['fraud_type'].value_counts())
```

## Visualizations

```
sns.countplot(data=redemptions, x='category', hue='is_fraud')
plt.title('Transactions by Category and Fraud Status')
plt.show()
```
